```python
def convert_str_num(string):
    if (type(string) == float) | (type(string) == np.int64) | (type(string) == np.float64):
        return float(string)
    if 'M' in string:
        num = float(string[:-1]) * 1e6
    elif 'B' in string:
        num = float(string[:-1]) * 1e9
    elif 'T' in string:
        num = float(string[:-1]) * 1e12
    elif '%' in string:
        num = float(string[:-1]) / 100
    else:
        num = float(string[:-1])
        
    return num
```

```python
tickers = ['crm', 'huya', 'fubo', 't', 'cost', 'pltr', 'sklz', 'vldr', 'goco', 'home', 'orbc', 
            'crsp', 'amzn', 'tsm', 'tsla', 'nio', 'eh', 'aapl', 'nvda', 'nndm', 'sqns', 'mass', 
            'veev', 'mga', 'tdy', 'lmt', 'expc', 'beke', 'nvs', 'rptx', 'api', 'hon', 'lgvw', 
            'rhhby', 'googl', 'pton', 'srpt', 'kspn']   # 

index = [ticker.upper() for ticker in tickers]
df = pd.DataFrame(0, index=index, columns=columns)

for ticker in tickers:
    try:
        qsheet = get_balance_sheet(ticker, yearly=False)
        qsheet = qsheet.iloc[:,:2]
        qsheet.columns = ["Recent", "Previous"]
        ysheet = get_balance_sheet(ticker)
        
        df.loc[sticker.upper(), 'Quarterly Cash Growth'] = (convert_str_num(qsheet.loc[qsheet.index.str.contains('cash'), 'Recent'].values[0]) - convert_str_num(qsheet.loc[qsheet.index.str.contains('cash'), 'Previous'].values[0])) / (convert_str_num(qsheet.loc[qsheet.index.str.contains('cash'), 'Previous'].values[0]))
        df.loc[sticker.upper(), 'Cash'] = convert_str_num(qsheet.loc[qsheet.index.str.contains('cash'), 'Recent'].values[0])
    except ValueError:
        print('Error occurs in {}: {}'.format('get_balance_sheet', ticker))
    
    
    try: 
        stats_val = get_stats_valuation(ticker)
        stats_val = stats_val.iloc[:,:2]
        stats_val.columns = ["Attribute", "Recent"]
#         earnings = get_earnings(sticker)
#         cf = get_cash_flow(sticker)
#         cf = cf.iloc[:, :1]
#         cf.columns = ['Recent']

        df.loc[ticker.upper(), 'Market Cap'] = convert_str_num(stats_val.loc[stats_val.Attribute.str.contains('Market Cap'), 'Recent'].values[0])
        df.loc[ticker.upper(), 'Enterprise Value'] = convert_str_num(stats_val.loc[stats_val.Attribute.str.contains('Enterprise Value'), 'Recent'].values[0])
        df.loc[ticker.upper(), 'PEG Ratio'] = convert_str_num(stats_val.loc[stats_val.Attribute.str.contains('PEG Ratio'), 'Recent'].values[0])
        df.loc[ticker.upper(), 'Price/Book'] = convert_str_num(stats_val.loc[stats_val.Attribute.str.contains('Price/Book'), 'Recent'].values[0])
        df.loc[ticker.upper(), 'Price/Sales'] = convert_str_num(stats_val.loc[stats_val.Attribute.str.contains('Price/Sales'), 'Recent'].values[0])
        df.loc[ticker.upper(), 'Forward P/E'] = convert_str_num(stats_val.loc[stats_val.Attribute.str.contains('Forward P/E'), 'Recent'].values[0])
        df.loc[ticker.upper(), 'Trailing P/E'] = convert_str_num(stats_val.loc[stats_val.Attribute.str.contains('Trailing P/E'), 'Recent'].values[0])
    except ValueError:
        print('Error occurs in {}: {}'.format('get_stats_valuation', ticker))
    
    
    try:
        stats = get_stats(ticker)
        stats = stats.dropna(how='all')
        df.loc[ticker.upper(), 'Shares Outstanding'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Shares Outstanding'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Quarterly Revenue Growth'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Quarterly Revenue Growth'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Average Dividend Yield'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Average Dividend Yield'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Held by Insiders'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Held by Insiders'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Held by Institutions'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Held by Institutions'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Gross Profit'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Gross Profit'), 'Value'].values[0])
        df.loc[ticker.upper(), 'EBITDA'] = convert_str_num(stats.loc[stats.Attribute.str.contains('EBITDA'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Quarterly Earnings Growth'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Quarterly Earnings Growth'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Total Cash'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Total Cash'), 'Value'].values[0])
        df.loc[ticker.upper(), 'Diluted EPS'] = convert_str_num(stats.loc[stats.Attribute.str.contains('Diluted EPS'), 'Value'].values[0])
    except ValueError:
        print('Error occurs in {}: {}'.format('get_stats', ticker))
    
df.to_csv('report/first_group.csv')
df
```
